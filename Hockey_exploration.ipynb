```python
All_results_2019['Home_Win'] = All_results_2019.G < All_results_2019['G.1']
```

```python
def home_winner (row):
    if row['Home_Win'] == True:
        return 1
    if row['Away_Win'] == True:
        return 0
```

```python
def away_winner (row):
    if row['Home_Win'] == True:
        return 0
    if row['Away_Win'] == True:
        return 1
```

```python
All_results_2019.apply (lambda row: home_winner(row), axis=1);
```

```python
All_results_2019['Home_Win'] = All_results_2019.apply (lambda row: home_winner(row), axis=1)
```

```python
All_results_2019['Away_Win'] = All_results_2019.apply (lambda row: away_winner(row), axis=1)
```

```python
All_results_2019.rename(columns = {'G':'Away_Goals','G.1':'Home_Goals','Unnamed: 5':'OT_Yes','Att.':'Attendance'},inplace=True)
```

```python
def OT_winner (row):
    if row['OT_Yes'] == 'OT':
        return 1
    if row['OT_Yes'] == 'SO':
        return 2
    else:
        return 0
```

```python
All_results_2019['OT_Yes'] = All_results_2019.apply (lambda row: OT_winner(row), axis=1)
```

```python
Home_win_total = Home_win_total.drop(columns=['OT_Yes']).reset_index()
```

```python
merged_teams = pd.concat([Home_win_total, Home_2019], axis=1, sort=False)
```

```python
merged_teams = merged_teams.drop(columns=['Team','Season','GP'])
```

```python
merged_teams.dtypes;
```

```python
merged_teams = merged_teams.rename(columns={'Home':'Team_Name'})
```

```python
merged_teams['Team_Name'] = merged_teams.Team_Name.astype('str')
```

```python
merged_teams['Team_Name'] = merged_teams['Team_Name'].str.replace(' ', '_',3)
```

```python
merged_teams.head()
```

```python
team_names_dropped = merged_teams.drop(columns=['Team_Name'])
```

```python
team_names_dropped;
```

```python
X = team_names_dropped.loc[:, team_names_dropped.columns != 'Home_Win']

y = team_names_dropped.Home_Win # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
clf = DecisionTreeClassifier(max_depth=5)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
